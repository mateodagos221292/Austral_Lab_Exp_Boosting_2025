```r
### ============================
###  DATOS
### ============================

## ---- Grupo A ----
A_DART  <- c(65817000, 65237000, 66007000, 65217000, 63237000)
A_GBDT  <- c(64367000, 63627000, 62827000, 62787000, 65027000)
A_RF    <- c(56597000, 55937000, 55947000, 53787000, 53087000)

## ---- Grupo B ----
### Lote 1
B_DART_1 <- c(64692600, 63557100, 65345100, 65074800, 65058900)
B_GBDT_1 <- c(64042500, 63116700, 65102400, 64749900, 65058900)
B_RF_1   <- c(57155700, 58301100, 58597800, 56347200, 56912700)

### Lote 2
B_RF_2   <- c(57445200, 57527100, 57927600, 58289700, 57632700)
B_GBDT_2 <- c(63878100, 62589900, 64256100, 63957600, 63394200)
B_DART_2 <- c(64118400, 64833000, 64474500, 63958100, 65037900)

# Combinamos RF y GBDT porque ambos tienen 10 observaciones válidas
B_RF   <- c(B_RF_1, B_RF_2)
B_GBDT <- c(B_GBDT_1, B_GBDT_2)
B_DART <- c(B_DART_1 ,B_DART_2 )
```
